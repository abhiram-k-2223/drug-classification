```python
import pandas as pd

df = pd.read_csv('/content/drug200.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('Drug', axis=1)
y = df['Drug']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
class labelencoder:
  def __init__(self, columns):
    self.columns = columns

  def fit_transform(self, df):
    for col in self.columns:
      le = LabelEncoder()
      df[col] = le.fit_transform(df[col])
    return df
```

```python
x_train = labelencoder(['Sex']).fit_transform(x_train)
x_train.head()
```

```python
x_test = labelencoder(['Sex']).fit_transform(x_test)
```

```python
class ordinalencoder:
  def __init__(self, columns):
    self.columns = columns

  def fit_transform(self,df):
    from sklearn.preprocessing import OrdinalEncoder
    oe = OrdinalEncoder()
    df[self.columns] = oe.fit_transform(df[self.columns])
    return df
```

```python
x_train = ordinalencoder(['BP','Cholesterol']).fit_transform(x_train)
x_test = ordinalencoder(['BP','Cholesterol']).fit_transform(x_test)

x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=100000)
lr.fit(x_train, y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, pred)*100
score
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
```

```python
score = accuracy_score(y_test, pred)*100
score
```
